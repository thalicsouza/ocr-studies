## 4) Distance between characteres

```python
# using a function to calculate Euclidean Distance for two-dimension plane

def euclidean(q1,q2,p1,p2):
  result = ((q1 - p1) ** 2) + ((q2 - p2) ** 2)
  result = math.sqrt(abs(result))
  return result
```

```python
q1 = 40
p1 = 25
q2 = 70
p2 = 45
euclidean(q1,q2,p1,p2)
```

```python
# using a function to calculate Euclidean Distance for 3 or more-dimension plane

def euclidean_(q,p):
  result = math.dist(q,p)
  return result
```

```python
  q = (2, 5, 9)
  p = (3, 8, 12)
  euclidean_(q,p)
```
